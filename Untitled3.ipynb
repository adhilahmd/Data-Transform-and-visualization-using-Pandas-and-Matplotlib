```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/retail_sales.csv')
```

```python
df
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
total_sales_per_product=df.groupby('Product')['Sales'].sum()
print(total_sales_per_product)
```

```python
bestselling_product=total_sales_per_product.sort_values(ascending=False).index[0]
print(bestselling_product)
```

```python
average_daily_sales=df.groupby('Date')['Sales'].mean()
print(average_daily_sales)
```

```python
class retailsales:
  def __init__(self) -> None:
    self.df=pd.read_csv('/content/retail_sales.csv')
    self.df['Date']=pd.to_datetime(df['Date'])
    self.df.dropna(inplace=True)

  def total_sales_per_product(self):
    return self.df.groupby('Product')['Sales'].sum()

  def bestselling_product(self):
    return total_sales_per_product.sort_values(ascending=False).index[0]

  def average_daily_sales(self):
    return self.df.groupby('Date')['Sales'].mean()

  def plot_sales_trend(self):
    self.df.groupby('Date')['Sales'].sum().plot(kind='line')
    plt.xlabel('Date')
    plt.ylabel('Sales')
    plt.show()

  def plot_total_Sales_per_product(self):
    self.total_sales_per_product().plot(kind='bar')
    plt.xlabel('Product')
    plt.ylabel('Sales')
    plt.show()

  def plot_average_daily_sales(self):
    self.average_daily_sales().plot(kind='bar')
    plt.xlabel('Date')
    plt.ylabel('Sales')
    plt.show()


```

```python
Analyze=retailsales()
print(Analyze.total_sales_per_product())
print(Analyze.bestselling_product())
print(Analyze.average_daily_sales())
print(Analyze.plot_sales_trend())
print(Analyze.plot_total_Sales_per_product())
print(Analyze.plot_average_daily_sales())
```
